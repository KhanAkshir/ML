```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import  make_classification
```

```python
X,y1=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y1
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y1,test_size=0.33,random_state=42)
```

```python
X_train
```

```python

y_train
```

```python
sns.scatterplot(x=pd.DataFrame(X_train)[0],y=pd.DataFrame(X_train)[1],hue=y_train)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Regressor (SVR)

from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=1000,n_features=2,n_targets=1,noise=3.0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR(kernel='linear')
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
y_pred
```

```python
    
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['linear'],
              'epsilon':[0.1,0.2,0.3]
              }
```

```python
grid=GridSearchCV(SVR(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(r2_score(y_test,y_pred4))
```
