# SMOTE

```python
from sklearn.datasets import make_classification 
```

```python
X,Y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
# X independent features
#Y dependent feature
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(Y,columns=['target'])
```

```python
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df.target.value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
```

```python
oversample=SMOTE() #initializing smote
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target']) #oversampling
```

```python
oversampled_final_df=pd.concat([X,y],axis=1)
```

```python
oversampled_final_df
```
