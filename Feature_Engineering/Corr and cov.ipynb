```python
# Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. 
# Suppose you have collected data on the amount of time students spend studying for an exam and their 
# final exam scores. Calculate the Pearson correlation coefficient between these two variables 
# and interpret the result.

study_time = [2, 4, 6, 8, 10]
exam_score = [70, 75, 80, 85, 90]
d={'ST':study_time , 'ES':exam_score}
df=pd.DataFrame(data=d)
df
```

```python
df=df.corr(method='pearson')
```

```python
list(df.ES)[0]
```

```python
# Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
# exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
# for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
# between these two variables and compare the results.



hours_of_exercise = [4, 6, 8, 5, 2, 1, 3, 7, 9, 6, 3, 2, 1, 5, 8, 7, 9, 6, 4, 3,
                     2, 1, 6, 8, 7, 9, 5, 4, 3, 2, 1, 6, 8, 7, 9, 5, 4, 3, 2, 1,
                     6, 8, 7, 9, 5, 4, 3, 2, 1]


BMI = [22.4, 24.5, 26.1, 23.5, 27.3, 30.5, 25.2, 24.3, 21.7, 23.6, 28.2, 29.7,
       32.6, 25.8, 23.7, 26.1, 21.8, 24.9, 26.5, 29.8, 30.1, 33.2, 27.7, 25.4,
       28.1, 32.4, 28.5, 29.6, 26.3, 24.5, 22.9, 27.2, 30.5, 31.7, 33.5, 29.1,
       28.4, 27.6, 26.4, 24.7, 22.8, 25.2, 27.6, 31.3, 29.4, 30.1, 28.8, 25.7,
       23.6]

```
