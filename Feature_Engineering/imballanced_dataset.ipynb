```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
n_sample=1000;
class_0_ratio=0.9
n_class_0=int(n_sample*class_0_ratio)
n_class_1=n_sample-n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0=pd.DataFrame(
{
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame(
{
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})

# making two dfs
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)  # kind of joining bith dataframes
```

```python
df.head()
```

```python
df.target.value_counts() 
```

## Upsampling

```python
df_minority=df[df['target']==1]
df_majority=df[df.target==0]
```

```python
df_majority
```

```python
df_minority
```

```python
# upsampling perform

from sklearn.utils import resample
```

```python
df_majority_upsample=resample(df_majority,replace=True,n_samples=len(df_majority))  # umsampling
```

```python
df_majority_upsample.shape
```

```python
df_upsample=pd.concat([df_majority,df_majority_upsample]).reset_index(drop=True)
```

```python
df_upsample
```

# Downsampling

```python
class_0=pd.DataFrame(
{
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame(
{
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority
```

```python
df_minority
```

```python
df_majority_downsample=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsample.shape
```

```python
pd.concat([df_minority,df_majority_downsample]).reset_index(drop=True)
```
