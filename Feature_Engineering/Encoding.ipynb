# OHE/Nolminal Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'color':['red','blue','green','red','blue']})
```

```python
df
```

```python
# create an instance for ohe
a=OneHotEncoder()
```

```python
ans=a.fit_transform(df[['color']]) 
ans
```

```python
encoded_df=pd.DataFrame(ans.toarray(),columns=a.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```

# Target Guided Ordinal Encoding

```python
df=pd.DataFrame({
    'city':['Newyork','London','Paris','Dubai','Kabul','Kolkata','Kabul'],
    'price':[200,150,300,250,180,320,500]
})
```

```python
df
```

```python
#claculate the mean price

mean_price=df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
# replace each city with its maean price
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```

```python
df
```

```python
median_price=df.groupby('city')['price'].median().to_dict()
median_price
```

```python
df['median_encode']=df['city'].map(median_price)
```

```python
df
```
