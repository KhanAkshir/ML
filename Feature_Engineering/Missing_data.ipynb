
# Imputatiom

### a).mean value imputation - this technique works well when your data is normally distributed

```python
sns.distplot(df.age)
```

```python
import numpy as np
df['Age_mean']=df['age'].fillna(np.mean(df['age']))
```

```python
df
```

```python
df
```

```python
sns.heatmap(df.isnull()) # here we can see that in age_mean all the null values are replaced
```

```python
df['Age_median']=df.age.fillna(df.age.median())
```

```python
df[df.embarked.isnull()]
```

```python
sns.heatmap(df.isnull())
```
