# Data interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

```

## 1).Linear Interpolation

```python
x_new=np.linspace(1,5,10)
```

```python
y_interpolation_point=np.interp(x_new,x,y)
```

```python
y_interpolation_point
```

```python
x_new
```

```python
plt.scatter(x_new,y_interpolation_point) # after interpolation
```

## 2). Cubic interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])

```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic') # creating a cubic interpolation func
```

```python
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

##  3).Polynomial Interpolatiom

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,16,66,64,125])
```

```python
p=np.polyfit(x,y,3)
```

```python
p
```

```python
x_new=np.linspace(1,5,20)
y_interp=np.polyval(p,x_new)
```

```python
plt.scatter(x_new,y_interp)
```
