# OHE/Nolminal Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'color':['red','blue','green','red','blue']})
```

```python
df
```

```python
# create an instance for ohe
a=OneHotEncoder()
```

```python
ans=a.fit_transform(df[['color']]) 
```

```python
encoded_df=pd.DataFrame(ans.toarray(),columns=a.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```
